```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("neuralchemy/Prompt-injection-dataset", "core")
train = ds["train"]
df = train.to_pandas()

print("Shape:", df.shape)
print("\nLabel counts:\n", df['label'].value_counts())
print("\nFirst few rows:\n", df.head())
```

```python
# Check category breakdown
print("Categories:\n", df['category'].value_counts())
print("\nSeverity:\n", df['severity'].value_counts())

# Save to your dataset folder
df.to_csv("../dataset/prompt_injection_dataset.csv", index=False)
print("\nSaved to dataset/prompt_injection_dataset.csv")
```
